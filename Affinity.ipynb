```python
import requests

def check_by_pincode(address, pincode):
    api_url = f'https://api.postalpincode.in/pincode/{pincode}'

    try:
        response = requests.get(api_url)
        data = response.json()

        if response.status_code == 200 and data and data[0]['Status'] == 'Success':
            # Check if any of the post offices match the provided address
            for post_office_info in data[0]['PostOffice']:
                post_office = post_office_info.get('Name', '').lower()
                state = post_office_info.get('State', '').lower()

                if post_office in address.lower() and state in address.lower():
                    return True

    except Exception as e:
        print(f"An error occurred: {e}")

    return False

# Test cases with correct data
correct_address_1 = "2nd Phase, 374/B, 80 Feet Rd, Mysore Bank Colony, Banashankari 3rd Stage, Srinivasa Nagar, Bengaluru, Karnataka 560050"
pincode_1 = "560050"

correct_address_2 = "2nd Phase, 374/B, 80 Feet Rd, Bank Colony, Banashankari 3rd Stage, Srinivasa Nagar, Bengaluru, Karnataka 560050"
pincode_2 = "560050"

correct_address_3 = "374/B, 80 Feet Rd, State Bank Colony, Banashankari 3rd Stage, Srinivasa Nagar, Bengaluru,Karnataka 560050"
pincode_3 = "560050"

correct_address_4 = "D-102 Golden Residency, Adarsh Nagar, Ganeshpur, Roorkee City, Roorkee, Uttarakhand 247667"
pincode_4 = "247667"

# Test cases with incorrect data
incorrect_address_1 = "2nd Phase, 374/B, 80 Feet Rd, Mysore Bank Colony, Banashankari 3rd Stage, Srinivasa Nagar, Bangalore, 560050"
pincode_5 = "560050"

incorrect_address_2 = "Colony, Bengaluru, Karnataka 560050"
pincode_6 = "560050"

# Test cases
print(f"Address 1 is valid by PIN code: {check_by_pincode(correct_address_1, pincode_1)}")  # Should return True
print(f"Address 2 is valid by PIN code: {check_by_pincode(correct_address_2, pincode_2)}")  # Should return True
print(f"Address 3 is valid by PIN code: {check_by_pincode(correct_address_3, pincode_3)}")  # Should return True
print(f"Address 4 is valid by PIN code: {check_by_pincode(correct_address_4, pincode_4)}")  # Should return True

print(f"Address 5 is valid by PIN code: {check_by_pincode(incorrect_address_1, pincode_5)}")  # Should return False
print(f"Address 6 is valid by PIN code: {check_by_pincode(incorrect_address_2, pincode_6)}")  # Should return False
```
